```python
import math
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def f_x(x):
    return 1/(2+2*x+x**2)
```

```python
def lagrange1(n, x):
    a = []
    b = []
    for j in range(n + 1):
        x0=-5 + 10.0 * j / n
        a.append(x0)
        b.append(f_x(x0))
    f = 0.0
    for i in range(n + 1):
        tmp = 1.0
        for j in range(i):
            tmp = tmp * (x - a[j]) / (a[i] - a[j])
        for j in range(i + 1, n + 1):
            tmp = tmp * (x - a[j]) / (a[i] - a[j])
        f = f + tmp * b[i]
    return f
```

```python
def lagrange2(n, x):
    c = []
    d = []
    for j in range(n + 1):
        x0=-5 * (math.cos((2 * j + 1) * math.pi / (2 * n + 2)))
        c.append(x0)
        d.append(f_x(x0))
    f = 0.0
    for i in range(n + 1):
        tmp = 1.0
        for j in range(i):
            tmp = tmp * (x - c[j]) / (c[i] - c[j])
        for j in range(i + 1, n + 1):
            tmp = tmp * (x - c[j]) / (c[i] - c[j])
        f = f + tmp * d[i]
    return f
```

```python
x_point=np.array([(i/50-5) for i in range(501)],dtype='float64')
y1=np.array([f_x(k) for k in x_point])
```

```python
def compute_error(type,n):
    x=0;
    if type==1:
        for i in range(501):
            x=max(math.fabs(lagrange1(n, x_point[i]) - f_x(x_point[i])), x)
    else:
        for i in range(501):
            x=max(math.fabs(lagrange2(n, x_point[i]) - f_x(x_point[i])), x)
    return x
```

```python
print("------------n=4------------")
print('1:  '+str(compute_error(1,4))+'   2:  '+str(compute_error(2,4)))  
y2=np.array([lagrange1(4, k) for k in x_point])
y3=np.array([lagrange2(4, k) for k in x_point])
plt.subplot(3,  1,  1)  
plt.title("fx")
plt.plot(x,y1)
plt.subplot(3,  1,  2)
plt.title("lagrange1")
plt.plot(x,y2)
plt.subplot(3,  1,  3)
plt.title("lagrange1")
plt.plot(x,y3)
plt.show()
```
